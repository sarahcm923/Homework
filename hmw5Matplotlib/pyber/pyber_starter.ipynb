```python
%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
#city_group = city_data_df.groupby(["city"])
#ride_group = ride_data_df.groupby(["city"])

# Combine the data into a single dataset
all_group = pd.merge(city_data_df, ride_data_df, on="city")

# Display the data table for preview
all_group.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#Urban
urban_df = all_group[all_group["type"] == "Urban"]
city_urban_df = urban_df.groupby(["city"])
x_axis_U = city_urban_df["ride_id"].count()
y_axis_U = city_urban_df["fare"].mean()
z_U = city_urban_df["driver_count"].sum()


#Suburban
suburb_df = all_group[all_group["type"] == "Suburban"]
city_suburb_df = suburb_df.groupby(["city"])
x_axis_S = city_suburb_df["ride_id"].count()
y_axis_S = city_suburb_df["fare"].mean()
z_S = city_suburb_df["driver_count"].sum()

#Rural
rural_df = all_group[all_group["type"] == "Rural"]
city_rural_df = rural_df.groupby(["city"])
x_axis_R = city_rural_df["ride_id"].count()
y_axis_R = city_rural_df["fare"].mean()
z_R = city_rural_df["driver_count"].sum()

# Build the scatter plots for each city types
plt.scatter(x_axis_U, y_axis_U, s=z_U, c="coral",edgecolors="black", alpha=0.6, linewidth=1, label = "Urban")
plt.scatter(x_axis_S, y_axis_S, s=z_S, c="skyblue",edgecolors="black", alpha=0.6, linewidth=1, label = "Suburban")
plt.scatter(x_axis_R, y_axis_R, s=z_R, c="gold",edgecolors="black", alpha=0.6, linewidth=1, label = "Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
# Create a legend
lgnd = plt.legend(loc="upper right", scatterpoints=1, fontsize=10, title = "City Types")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
# Incorporate a text label regarding circle size

plt.gcf().text(.92, 0.5, "Note:\nCircle size correlates with driver county per city.", fontsize=10)
# Save Figure
plt.savefig("data/bubble chart output.png", bbox_inches="tight")
plt.close()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_U = urban_df["fare"].sum()
fare_S = suburb_df["fare"].sum()
fare_R = rural_df["fare"].sum()

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
sizes = [fare_U, fare_S, fare_R]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)
plt.title('% of Total Fares by City Type')
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=270)

# Save Figure
plt.savefig("data/fares by city type.png")
plt.close()
```

```python
# Show Figure
plt.show()
```
